```python
# Subset the relevant columns from the movies dataset
subset = movies[['awards.wins', 'countries']]

# Create an empty dictionary to store country wins
country_wins = {}

# Iterate over the rows of the subset DataFrame
for index, row in subset.iterrows():
    countries = row['countries']
    wins = row['awards.wins']
    
    # Check if there are multiple countries in the row
    if isinstance(countries, list):
        num_countries = len(countries)
        
        # Assign the award wins to each country individually
        for country in countries:
            if country in country_wins:
                country_wins[country] += wins / num_countries
            else:
                country_wins[country] = wins / num_countries
    else:
        # Assign the award wins to the single country
        if countries in country_wins:
            country_wins[countries] += wins
        else:
            country_wins[countries] = wins

# Convert the country wins dictionary to a DataFrame
country_wins_df = pd.DataFrame.from_dict(country_wins, orient='index', columns=['Total Wins'])
country_wins_df.index.name = 'Country'

# Sort the data in descending order and select the top 15 countries
top_countries = country_wins_df.nlargest(15, 'Total Wins')

# Calculate the sum of wins for the countries not in the top 15
other_wins_sum = country_wins_df.loc[~country_wins_df.index.isin(top_countries.index), 'Total Wins'].sum()

# Add the sum of wins for other countries as the 15th entry
top_countries.loc['Other'] = other_wins_sum

# Plot the bar plot
plt.figure(figsize=(12, 6))
top_countries.plot(kind='bar')
plt.xlabel('Countries')
plt.ylabel('Total Award Wins')
plt.title('Number of Award Wins by Countries')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a copy of the movies dataset
movies_copy = movies.copy()

# Split the movies with multiple countries into separate rows
movies_copy['countries'] = movies_copy['countries'].apply(lambda x: x.split(',') if isinstance(x, str) else x)
movies_copy = movies_copy.explode('countries')

# Remove duplicate entries
movies_copy = movies_copy.drop_duplicates(subset=['_id', 'awards.wins', 'countries'])

# Preprocess country names to remove unwanted characters
# During the creation of this chart, I had a duplicate country error in which countries would reappear many times
# For example, "USA", ["USA", and "USA"] were all in the top 15.
# To solve this, I preprocessed the country names by stripping any leading or trailing whitespaces and removing square brackets 
# using the .strip() and .strip('[]') string methods. This helps eliminate any unwanted characters that caused duplicates.
movies_copy['countries'] = movies_copy['countries'].str.strip().str.strip('[]')

# Subset the relevant columns from the movies copy dataset
subset = movies_copy[['awards.wins', 'countries']]

# Create an empty dictionary to store country wins
country_wins = {}

# Iterate over the rows of the subset DataFrame
for index, row in subset.iterrows():
    country = row['countries']
    wins = row['awards.wins']
    
    # Assign the award wins to each country individually
    if country in country_wins:
        country_wins[country] += wins
    else:
        country_wins[country] = wins

# Convert the country wins dictionary to a DataFrame
country_wins_df = pd.DataFrame.from_dict(country_wins, orient='index', columns=['Total Wins'])
country_wins_df.index.name = 'Country'

# Sort the data in descending order and select the top 15 countries
top_countries = country_wins_df.nlargest(15, 'Total Wins')

# Calculate the sum of wins for the countries not in the top 15
other_wins_sum = country_wins_df.loc[~country_wins_df.index.isin(top_countries.index), 'Total Wins'].sum()

# Add the sum of wins for other countries as the 15th entry
top_countries.loc['Other'] = other_wins_sum

# Plot the bar plot
plt.figure(figsize=(12, 6))
top_countries.plot(kind='bar')
plt.xlabel('Countries')
plt.ylabel('Total Award Wins')
plt.title('Number of Award Wins by Countries (Top 15)')
plt.xticks(rotation=45)

# Adjust the spacing of the x-axis labels
plt.tight_layout()

plt.show()
```

```python
# Handle missing values
movies_2 = movies.dropna(subset=['imdb.rating', 'genres'])

# Split genres
genres_df = movies_2.assign(genres=movies_2['genres'].str.split(', ')).explode('genres')

# Aggregate data
aggregated_df = genres_df.groupby('genres').agg({'imdb.rating': 'mean', 'awards.wins': 'sum'}).reset_index()

# Sort values
aggregated_df = aggregated_df.sort_values(by='imdb.rating', ascending=False)

# Plot the data
fig, ax1 = plt.subplots()

# Plot average IMDB rating
ax1.bar(aggregated_df['genres'], aggregated_df['imdb.rating'], color='b')
ax1.set_xlabel('Genre')
ax1.set_ylabel('Average IMDB Rating', color='b')
ax1.tick_params('y', colors='b')

# Create a second y-axis for total award wins
ax2 = ax1.twinx()
ax2.plot(aggregated_df['genres'], aggregated_df['awards.wins'], color='r', marker='o')
ax2.set_ylabel('Total Award Wins', color='r')
ax2.tick_params('y', colors='r')

plt.title('Average IMDb Rating and Total Award Wins by Genre')
plt.xticks(rotation=90)
plt.show()
```

```python
# Handle missing values
movies = movies.dropna(subset=['imdb.rating', 'genres'])

# Create a copy of the DataFrame
movies_2 = movies.copy()

# Split genres
movies_2.loc[:, 'genres'] = movies_2['genres'].str.split(', ')

# Explode genres (Transform each element of a list-like to a row, replicating index values.)
movies_2 = movies_2.explode('genres')

# Aggregate data
aggregated_df = movies_2.groupby('genres').agg({'imdb.rating': 'mean', 'awards.wins': 'sum'}).reset_index()

# Select top N genres based on total award wins
aggregated_df = aggregated_df.sort_values(by='awards.wins', ascending=False)
top_genres = aggregated_df['genres'].head(20).tolist()

# Calculate the sum of all other genres
other_wins = aggregated_df[~aggregated_df['genres'].isin(top_genres)]['awards.wins'].sum()
other_average_rating = aggregated_df[~aggregated_df['genres'].isin(top_genres)]['imdb.rating'].mean()

# Create a new DataFrame for visualization
visual_df = pd.DataFrame({
    'genres': top_genres + ['Other'],
    'imdb.rating': aggregated_df.loc[aggregated_df['genres'].isin(top_genres), 'imdb.rating'].tolist() + [other_average_rating],
    'awards.wins': aggregated_df.loc[aggregated_df['genres'].isin(top_genres), 'awards.wins'].tolist() + [other_wins]
})

# Sort values
visual_df = visual_df.sort_values(by='imdb.rating', ascending=False)

# Plot the data
fig, ax1 = plt.subplots(figsize=(10, 6))  # Adjust the figure size

# Plot average IMDb rating as individual dots
ax1.scatter(visual_df.index, visual_df['imdb.rating'], color='blue', label='Average IMDb Rating', zorder=10)

# Plot total award wins as bars
ax2 = ax1.twinx()
ax2.bar(visual_df.index, visual_df['awards.wins'], color='red', label='Total Award Wins')

# Set x-axis labels
ax1.set_xticks(visual_df.index)
ax1.set_xticklabels(visual_df['genres'], rotation=45, ha='right')

# Set y-axis labels and legends
#ax1.set_ylabel('Average IMDb Rating')
#ax2.set_ylabel('Total Award Wins')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Extend the y-axis range
ax1.set_ylim(bottom=visual_df['imdb.rating'].min() - 0.5, top=visual_df['imdb.rating'].max() + 0.5)

# Adjust the spacing between the plot and the legend
plt.subplots_adjust(right=1.1)  # Adjust the right parameter to increase or decrease the spacing

plt.title('Average IMDb Rating and Total Award Wins by Genre/Genre Pairs')
plt.tight_layout()
plt.show()
```
